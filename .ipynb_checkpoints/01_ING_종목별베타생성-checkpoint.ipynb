## 시장 베타 계수  구하기

```python
import numpy as np 
from statsmodels import regression 
import statsmodels.api as sm 
import matplotlib.pyplot as plt 
import math 
import pandas as pd 
import pandas_datareader as pdr
```

```python
# 삼성전자
stock_df = fdr.DataReader('005930','2017-10-10','2018-10')
# stock_df['Close'].plot()
stock_df
```

```python
# import FinanceDataReader as fdr

# KS11 (KOSPI 지수), 
ks_df = fdr.DataReader('KS11','2017-10-11','2018-10') # 삼성전자 171010데이터 오류
ks_df['Close']
ks_df.count()
```

```python
merged_df = pd.merge(stock_df, ks_df, left_index=True,right_index=True)
merged_df
# merged_df.to_csv('./datas/merged.csv')
```

```python
def linreg(x,y): 
    x = sm.add_constant(x) 
    model = regression.linear_model.OLS(y,x).fit() 
    x = x[:,1] 
    return model.params[0], model.params[1] 

X = list(merged_df['Change_y'])
Y = list(merged_df['Change_x']) # 지수
```

```python
len(X)
```

```python

alpha, beta = linreg(X,Y) 
print('alpha:' + str(alpha)) 
print('beta:' + str(beta))
```
