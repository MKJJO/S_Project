## 6. 첨도, 왜도 함수 생성

https://datascienceschool.net/view-notebook/8c39a71b453e43d9bd1000f38638e937/

https://en.wikipedia.org/wiki/Algorithms_for_calculating_variance#On-line_algorithm


```python
# 첨도, 왜도 함수
def kur_ske_cal(factor, data,stat):
    mean = stat['mean']
    mean = mean.values[0]
    k_sum1 = k_sum2 = s_sum1 = s_sum2 = 0
    n = len(data)

    for x in data:

        # 첨도용
        k_sum1 = k_sum1 + (x-mean)**4 # 거듭제곱
        k_sum2 = k_sum2 + (x-mean)**2

        # 왜도용
        s_sum1 = s_sum1 + (x-mean)**3 # 거듭제곱
        s_sum2 = s_sum2 + (x-mean)**2


        #print("error :",factor, "  value :", x, "  mean :", mean, s_sum1,s_sum2)
            
    kurtosis = (k_sum1/n) / ((k_sum2/n)**2) - 3    
    skewness = (s_sum1/n) / ((s_sum2/n)**(3/2))
    return kurtosis, skewness

```

## 7. 정규화 원칙

### 1) 전처리 함수 생성

#### 아웃라이어 제거 함수

```python
# 아웃라이어(6시그마) 제거 함수

def outlier_remove(data, stat):
    
    mean = stat['mean']
    mean = mean.values[0]
    min = stat['min']
    min = min.values[0]
    max = stat['max']
    max = max.values[0]
    stdev = stat['stdev']
    stdev = stdev.values[0]
    
    #print("mean", mean,"stdev",stdev, "min", min, "max", max )
    
    result = []
    
    for x in data:
        if x < mean and x < (min + 3*stdev):
            x = (min + 3*stdev)
        elif x > mean and x > (max - 3*stdev):
            x = (max - 3*stdev)
        else: x = x
        
    
        result.append(x)
        
    return result
    
```

#### 표준정규화 함수

```python
# 표준정규화
# (x-mean)/stdev

def standardization(data, stat):
    
    mean = stat['mean']
    mean = mean.values[0]
    min = stat['min']
    min = min.values[0]
    max = stat['max']
    max = max.values[0]
    stdev = stat['stdev']
    stdev = stdev.values[0]
    
    #print("mean", mean,"stdev",stdev, "min", min, "max", max )
    
    result = []
    
    for x in data:
        x = (x-mean)/stdev      
        result.append(x)
        
    return result
    
```

#### 5Class 카테고리화 함수

```python
# 공시, 적자누적 횟수 등의 항목은 5개 분류 처리(단, max가 5초과할 경우, 0->0(자체에 의미 있음) 1이상값에 대해 분류(1~4, 4개))


def class_categories(data, stat):
    
    mean = stat['mean']
    mean = mean.values[0]
    min = stat['min']
    min = min.values[0]
    max = stat['max']
    max = max.values[0]
    stdev = stat['stdev']
    stdev = stdev.values[0]
    
    #print("mean", mean,"stdev",stdev, "min", min, "max", max )
    
    # 0을 제외한 퍼센타일 구하기
    temp = pd.DataFrame()
    temp['value'] = data
    temp = temp[temp['value'] != 0]
    p25 = np.percentile(temp, 25)
    p50 = np.percentile(temp, 50)
    p75 = np.percentile(temp, 75)
    
    result = []
    
    for x in data:
        if max > 5 and x != 0:
            # max가 5이하 이거나, 이상이더라도 0은 0값 그대로 사용
            if x <= p25:
                x = 1
            elif x <= p50:
                x = 2
            elif x <= 75:
                x = 3
            else: x = 4
            
        result.append(x)
        
    return result
    
```

### 2) 정규화 일괄 적용

```python
stat1 = data_stat[data_stat.key == 'EN01174011']
stat2 = stats_df[stats_df.key == 'EN01174011']
value = stat2['key']
value.values[0]
```

```python
result_df = data
method_df = pd.DataFrame()
method_list = []
factors = []

for x in cols[3:]:
    
    # 해당 factor 통계치 가져오기
    stat1 = data_stat[data_stat.key == x] # 평균, 중앙값, min, max, 편차
    stat2 = stats_df[stats_df.key == x]  #첨도 왜도 
    type = stat1['type']
    type = type.values[0]    
    size = stat1['size']
    size = size.values[0]       
    min = stat1['min']
    min = min.values[0] 
    max = stat1['max']
    max = max.values[0] 
    skewness = stat2['skewness']
    skewness = skewness.values[0]   
    add = 0
    method = ''
    
    try:

        if abs(skewness) > 3 and type == 'amount' and size != 'small':
            # step1 조건 : 왜도 절대값이 3 초과, 타입이 amount, 규모가 not small
            if min < 0:
                add = min*-1   # log처리 전 음수 양수화(+ min + 1)
            else:
                add = 0
            result_df[x] = result_df[x].map(lambda z : z+add)
            result_df[x].replace(0, 1, inplace=True)
            result_df[x] = result_df[x].map(lambda z : np.log(z))
            print("log처리 : ", x)
            method = 'log'
        elif type == 'count':
            # step2 조건 : 타입이 count
            if max != 0:
                result_df[x] = class_categories(result_df[x], stat1)
                print("5class 카테고리 : ", x)
                method = '5class'
        else:
            # 나머지 step3,4
            outlier_removed = outlier_remove(result_df[x], stat1)
            result_df[x] = standardization(outlier_removed, stat1)
            print("표준정규화 : ", x)
            method = 'standard'
    except:
        print("error : ", x)
        method = 'error'
    
    method_list.append(method)
    factors.append(x)
    
    
method_df['key'] = factors
method_df['method'] = method_list
```
